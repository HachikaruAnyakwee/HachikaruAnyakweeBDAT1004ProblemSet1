````python
var values = new object[]
{
    5,
    5.0,
    5 > 1,
    "5",
    5 * 2,
    "5" + "2",
    "5" + 2,
    5.0 / 2,
    5 % 2,
    new HashSet<int> { 5, 2, 1 },
    Math.PI
};

foreach (var value in values)
{
    string dataType = GetDataType(value);
    Console.WriteLine($"The data type of {value} is {dataType}");
}

string GetDataType(object value)
{
    if (value is int) return "int";
    if (value is double) return "double";
    if (value is bool) return "bool";
    if (value is string) return "string";
    if (value is HashSet<int>) return "HashSet<int>";
    if (value is double && (double)value == Math.PI) return "Math.PI (double)";
    return "unknown";
}
````

````python
//Question 4 in C#

//Write a program in C# Sharp to separate odd and even integers in separate arrays.

//Go to the editor Test Data:

//Input the number of elements to be stored in the array :5 Input 5 elements in the array :

//element - 0 : 25
//element - 1 : 47
//element - 2 : 42
//element - 3 : 56
//element - 4 : 32 
//Expected Output:

//The Even elements are:
//42 56 32

//The Odd elements are :
//25 47
````

````python
using System;

// Define the number of elements
int n = 5;

// Initialize arrays to store even and odd numbers
int[] arr = { 25, 47, 42, 56, 32 };
int[] evenArr = new int[n];
int[] oddArr = new int[n];

// Initialize counters for even and odd elements
int evenCount = 0;
int oddCount = 0;

// Separate even and odd numbers
foreach (int num in arr)
{
    if (num % 2 == 0)
    {
        evenArr[evenCount] = num; // Store even numbers
        evenCount++;
    }
    else
    {
        oddArr[oddCount] = num; // Store odd numbers
        oddCount++;
    }
}

// Display even elements
Console.WriteLine("The Even elements are:");
for (int i = 0; i < evenCount; i++)
{
    Console.Write(evenArr[i] + " ");
}

// Display odd elements
Console.WriteLine("\nThe Odd elements are:");
for (int i = 0; i < oddCount; i++)
{
    Console.Write(oddArr[i] + " ");
}
````

````python
using System;

// Function to check if a point (x, y) is inside a rectangle defined by (x1, y1) and (x2, y2)
bool IsInsideRectangle(double x, double y, double x1, double y1, double x2, double y2)
{
    return x >= x1 && x <= x2 && y >= y1 && y <= y2;
}

// Perform tests and print results
void TestRectangleContainment(double x, double y, double x1, double y1, double x2, double y2, string testName)
{
    bool isInside = IsInsideRectangle(x, y, x1, y1, x2, y2);
    Console.WriteLine($"{testName} ({x}, {y}) is inside the rectangle ({x1}, {y1}, {x2}, {y2}): {isInside}");
}

// Test cases
TestRectangleContainment(1, 1, 0, 0, 2, 3, "a. (1, 1) is inside");
TestRectangleContainment(-1, -1, 0, 0, 2, 3, "a. (-1, -1) is inside");
TestRectangleContainment(1, 1, 0.3, 0.5, 1.1, 0.7, "b. (1, 1) is inside the first rectangle");
TestRectangleContainment(1, 1, 0.5, 0.2, 1.1, 2, "b. (1, 1) is inside the second rectangle");
````

````python
#Question 6 Python

#Pig Latin Transformation

#You can transform a word into Pig Latin using the following two rules:

#1. If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay,' and 'pencil' becomes 'encilpay.'
#2. If the word starts with a vowel, simply append 'way' to the end of the word. For example, 'enter' becomes 'enterway,' and 'other' becomes 'otherway.'

#For our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
#Write a function pig() that takes a word (i.e., a string) as input and returns its Pig Latin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case, however.
````

````python
def pig_latin(word):
    # Convert the word to lowercase
    word = word.lower()
    
    # Check if the word starts with a vowel
    vowels = 'aeiou'
    
    if word[0] in vowels:
        # If it starts with a vowel, append 'way' to the end
        return word + 'way'
    else:
        # If it starts with a consonant, move the first letter to the end and append 'ay'
        return word[1:] + word[0] + 'ay'

# Input from the user
user_input = input("Enter a word: ")

# Transform the word to Pig Latin and print the result
result = pig_latin(user_input)
print("Pig Latin transformation:", result)
````

````python
#Question 8:

#Currency Conversion

#Write a function curconv() that takes as input:

#1. A currency represented using a string (e.g., 'JPY' for the Japanese Yen or 'EUR' for the Euro).

#2. An amount in the given currency.

#The function should convert and return the amount in US dollars.

#Example Usage:
#curconv('EUR', 100)
#122.96544

#```python
#curconv('JPY', 100)
#1.241401


#The currency rates you will need for the conversion are provided below:

#AUD 1.0345157 Australian Dollar
#CHF 1.0237414 Swiss Franc
#CNY 0.1550176 Chinese Yuan
#DKK 0.1651442 Danish Krone
#EUR 1.2296544 Euro
#GBP 1.5550989 British Pound
#HKD 0.1270207 Hong Kong Dollar
#INR 0.0177643 Indian Rupee
#JPY 0.01241401 Japanese Yen
#MXN 0.0751848 Mexican Peso
#MYR 0.3145411 Malaysian Ringgit
#NOK 0.1677063 Norwegian Krone
#NZD 0.8003591 New Zealand Dollar
#PHP 0.0233234 Philippine Peso
#SEK 0.148269 Swedish Krona
#SGD 0.788871 Singapore Dollar
#THB 0.0313789 Thai Baht
````

````python
def read_currency_exchange_rates(file_name):
    # Create a dictionary to store currency exchange rates
    currency_dict = {}
    
    # Open the specified file and read exchange rate data
    with open(file_name, "r") as infile:
        for line in infile:
            parts = line.split()
            if len(parts) >= 2:
                currency_code = parts[0]
                exchange_rate = float(parts[1])
                currency_dict[currency_code] = exchange_rate
    
    return currency_dict

def curconv(currency_code, amount, exchange_rates):
    # Check if the currency code is in the dictionary
    if currency_code in exchange_rates:
        # Perform currency conversion and return the result
        converted_amount = amount * exchange_rates[currency_code]
        return converted_amount
    else:
        # Return an error message if the currency code is not found
        return "Currency code not found"

# Read currency exchange rates from 'currencies.txt'
exchange_rates = read_currency_exchange_rates('currencies.txt')

# Perform currency conversion and print the results
usd_value1 = curconv('EUR', 100, exchange_rates)
print(f"100 EUR in USD: {usd_value1:.2f}")

usd_value2 = curconv('JPY', 100, exchange_rates)
print(f"100 JPY in USD: {usd_value2:.2f}")
````

````python
#Question 10:

#Character Frequencies

#Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no one but the intended recipient can understand the encrypted message. Cryptanalysis refers to attempts to undo the encryption, even if some details of the encryption are unknown (for example, if an encrypted message has been intercepted). The first step of cryptanalysis is often to build up a table of letter frequencies in the encrypted text.
#Assume that the string letters is already defined as 'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies() that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text. Your function may ignore any characters that are not in letters.

#frequencies('The quick red fox got bored and went home.')
#[1, 1, 1, 3, 5, 1, 1, 2, 1, 0, 1, 0, 1, 2, 4, 0, 1, 2, 0, 2, 1, 0, 1, 1, 0, 0]

#```python
#frequencies('apple')
````

````python
def letter_frequencies(input_string):
    # Define the lowercase letters
    letters = 'abcdefghijklmnopqrstuvwxyz'
    
    # Initialize a list to store the frequencies of each letter
    frequency_list = [0] * len(letters)
    
    # Convert the input string to lowercase for consistent counting
    input_string = input_string.lower()
    
    for char in input_string:
        # Check if the character is a lowercase letter
        if char in letters:
            # Find the position of the letter in the alphabet
            index = letters.index(char)
            
            # Increment the frequency count for the corresponding letter
            frequency_list[index] += 1
    
    return frequency_list

# Calculate and print the letter frequencies
input_text = 'The quick red fox got bored and went home.'
frequencies = letter_frequencies(input_text)

for i in range(len(frequencies)):
    letter = chr(ord('a') + i)  # Convert index to corresponding letter
    count = frequencies[i]
    print(f"The letter '{letter}' appears {count} times.")
````
